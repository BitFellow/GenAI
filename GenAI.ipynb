```python
!pip install gensim

#Program 1
from gensim.models import Word2Vec

corpus = [
    ["king", "strong"],
    ["queen", "wise"],
    ["boy", "young"],
    ["girl", "young"],
    ["prince", "king"],
    ["princess", "queen"]
]

model = Word2Vec(corpus, vector_size=10, min_count=1)

# Computation
king_vec = model.wv["king"]
similarity = model.wv.similarity("king", "queen")
most_sim = model.wv.most_similar("king")

sum_vector = king_vec + model.wv["queen"]

diff_vector = king_vec - model.wv["queen"]

# Output (all prints at end)
print("Vector (king):", king_vec)
print("Similarity (king, queen):", similarity)
print("Most similar to king:", most_sim)

print("Vector Sum (king + queen):", sum_vector)
print("Vector Difference (king - queen):", diff_vector)
```

```python
#Program 4
from transformers import pipeline

analyzer = pipeline("sentiment-analysis")

sentences = [
    "I love AI",
    "This is bad"
]

results = []

# Processing
for s in sentences:
    res = analyzer(s)[0]
    results.append((s, res['label'], res['score']))

# Output (clean format like PDF)
for sentence, label, score in results:
    print("Sentence:", sentence)
    print("Sentiment:", label)
    print()
```

```python
#Program 6
from transformers import pipeline

model = pipeline("sentiment-analysis")

texts = [
    "Amazing product",
    "Very bad service"
]

results = []

# Processing
for t in texts:
    res = model(t)[0]

    # Create clean dictionary
    result_dict = {
        "Sentence": t,
        "Sentiment": res['label'],
        "Score": round(res['score'], 4)
    }

    results.append(result_dict)

# Output
for r in results:
    print(r)
```

```python
#Program 7
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

text = """Artificial intelligence is rapidly transforming modern industries by automating repetitive tasks, improving decision-making, and increasing overall efficiency.
In sectors such as healthcare, finance, and transportation, AI systems help analyze large volumes of data to identify patterns and make accurate predictions.
For example, AI can assist doctors in diagnosing diseases, help banks detect fraudulent transactions, and enable self-driving vehicles to operate safely.
Despite these advantages, there are concerns regarding job displacement, data privacy, and ethical use of AI technologies.
Organizations must ensure responsible implementation of AI while maximizing its benefits for society."""

# Generate SHORT output
result = generator(
    text,
    max_new_tokens=20,   # control size
    do_sample=False      # make it deterministic (no random long text)
)

summary = result[0]['generated_text']

# Keep only first sentence (simple trimming)
summary = summary.split(".")[0] + "."

print("Original Text:", text)
print("Summary:", summary)
```

```python
#Program 10
!pip install pymupdf -q

import fitz  # PyMuPDF

# Step 1: Read PDF content
def load_pdf(file):
    text = ""
    pdf = fitz.open(file)

    for page in pdf:
        text += page.get_text()

    return text.lower()

# Step 2: Chatbot
def chatbot():
    data = load_pdf("ipc.pdf")

    while True:
        query = input("Enter your query (type 'exit' to stop): ").lower()

        if query == "exit":
            print("Chatbot stopped.")
            break

        found = False

        # Search line by line
        for line in data.split("\n"):
            if query in line:
                print("Response:", line)
                found = True

        if not found:
            print("Response: No information found.")

# Run chatbot
chatbot()
```
